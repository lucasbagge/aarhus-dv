### b) Bestem vektoren altså bestem norm

Vi kan bruge numpy `np.linalg.norm` til dette formål

```python
x_y = np.array([x,y])
print(x_y)
```

```python
OP = np.linalg.norm(x_y)
print("Længden er:", '%.3g' % OP )
```
